# Cuaderno para realizar el EDA del data se de prediccion de manetnimiento en maquinaria pesada

### Diccionario de Datos

| No | Variable | Descripción |
| :--- | :--- | :--- |
| 1 | **UID** | Identificador único correlativo desde 1 hasta 10,000. |
| 2 | **product ID** | ID de variante de calidad; número de serie específico que incluye la letra de la variante. |
| 3 | **type** | Tipo de producto basado en el ID: **L** (Bajo - 50%), **M** (Medio - 30%) o **H** (Alto - 20%). |
| 4 | **air temperature [K]** | Temperatura del aire en Grados Kelvin (K); generada mediante *random walk* y normalizada a 300 K con una desviación estándar de 2 K. |
| 5 | **process temperature [K]** | Temperatura del proceso de manufactura en escala Kelvin [K]. |
| 6 | **rotational speed [rpm]** | Velocidad de rotación en rpm; calculada a partir de una potencia de 2860 W con ruido de distribución normal. |
| 7 | **torque [Nm]** | Par motor (torque) en Nm; distribuido normalmente alrededor de 40 Nm (DE = 10 Nm) sin valores negativos. |
| 8 | **tool wear [min]** | Desgaste de la herramienta; las variantes H/M/L añaden 5/3/2 minutos de desgaste respectivamente al proceso. |
| 9 | **machine failure** | **Variable objetivo:** Indica si la máquina falló (1) o no falló (0) durante el proceso. |
